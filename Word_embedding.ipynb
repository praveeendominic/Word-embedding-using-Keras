```python
import nltk
nltk.download()
```

```python
# STEPS:
# 1. Get sentences and set vocab size and max sentence length
# 2. Get Padded One hot encoding representation
# 3. Create the embedding matrix
```

```python
# 1. Get sentences and set vocab size and max sentence length
sentences=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good']

vocab_size=10000
max_sent_len=max([len(nltk.word_tokenize(x)) for x in sentences])
max_sent_len
```

```python
# 2. Get Padded One hot encoding representation

from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.preprocessing.sequence import pad_sequences

ohe_rep=[one_hot(x,vocab_size) for x in sentences]
print(f'***OHE***\n{ohe_rep}')
padded_ohe=pad_sequences(ohe_rep,padding='pre',maxlen=max_sent_len)
print(f'\n***PADDED OHE***\n{padded_ohe}')
```

```python
# 3. Create the embedding matrix
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding

model=Sequential()
model.add(Embedding(input_dim=vocab_size,output_dim=10,input_length=max_sent_len))
model.compile('adam','rmse')
print(model.summary())

print("\n****PREDICT****")
model.predict(padded_ohe)
```
